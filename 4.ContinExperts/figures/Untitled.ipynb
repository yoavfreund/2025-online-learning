```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define x values
x = np.linspace(0, 1, 500)

# Uniform prior (Beta(1, 1))
uniform_prior = beta.pdf(x, 1, 1)

# Dirichlet (1/2, 1/2) prior (equivalent to Beta(1/2, 1/2))
dirichlet_prior = beta.pdf(x, 0.5, 0.5)

# Plot both priors on the same graph
plt.figure(figsize=(10, 6))
plt.plot(x, uniform_prior, label="Uniform Prior (Beta(1, 1))", color="green")
plt.plot(x, dirichlet_prior, label="Dirichlet (1/2, 1/2) Prior", color="blue")
plt.title("Comparison of Uniform and Dirichlet (1/2, 1/2) Priors")
plt.xlabel("x")
plt.ylabel("Density")
plt.ylim([0,4])
plt.legend()
plt.grid(True)
plt.show()
```
